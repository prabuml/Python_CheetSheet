# This data structure or library is to read and analysis the csv/excel   

```python
import pandas as pd
```

```python
data = pd.read_csv('City_Zhvi_AllHomes.csv')
```

```python
# To get the all the columns in the data
data.columns
```

```python
# Head operation will display only the top values
data.head(5)
```

```python
# Tail operation will display only the top values
data.tail(5)
```

```python
# To get random sample subset of dataframe
data.sample(5)
```

```python
# To sample using fraction
data.sample(frac=0.1).head(5)
```

```python
# To select based on index
data.iloc[0:5]
```

```python
# To get single column
data['RegionID'].head(5)
```

```python
# To get multiple columns 
data[['RegionID', 'RegionName']].head()
```

```python
# To ge the count of each value in column
data['RegionName'].value_counts().head(5)
# Head operation will display only the top values
```

```python
# To get unique values of the column
data['RegionName'].unique(),data['RegionName'].nunique()
```

```python
# To get all the statistical value of numerical column
data.SizeRank.describe()
```

```python
# To select the particular rows in column
data[data['RegionName']=="Springfield"].head(5)
# Head operation will display only the required values
```

```python
# To select filter the list of row value in column
data[data['RegionName'].isin(["Springfield","New York"])].head(5)
# Head operation will display only the required values
```

```python
# To add if condition to filter the list
data[(data.SizeRank>100) & (data.SizeRank<150)].head()
```

```python
# To get all the top values of Size Rank
data.nlargest(5,'SizeRank')
```

```python
# To the get the max of the column. There are other builtin statisical function such as min, mean etc
data.SizeRank.max()
```

```python
# To define lamda function 
data.SizeRank.apply(lambda x: x*x).head(5)
```

```python
# To define function in apply
def square(x):
    return x*x
data.SizeRank.apply(square).head()
```

```python
# To define function using multiple column
def merge(row):
    return str(row["State"])+"_"+str(row["Metro"]) 
data.apply(merge,axis=1).head(5)
```

```python
data.apply(lambda row: str(row["State"])+"_"+str(row["Metro"]),axis=1).head(5)
```

```python
# To get groupby operation as similar to group by operation in sql list
data.groupby(['State'])['SizeRank'].min().head(5)
data.groupby(['State'])['SizeRank'].size().head(5)
```

```python
# To join operation as similar to jon operation in sql
sample_data = pd.read_csv('City_Zhvi_AllHomes_Sample.csv')
pd.merge(data[['RegionID','RegionName']],sample_data,how='inner',on=['RegionID'])
```

```python
pd.merge(data[['RegionID','RegionName']],sample_data,how='left',on=['RegionID']).head(20)
```
