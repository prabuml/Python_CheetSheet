Sets — 
<br>
&emsp; Unordered collections of unique elements¶

```python
from IPython.display import Image 
```

```python
# Creating a set

a={0,1,2}
b = {3,0}
```

```python
# To search element in a list
0 in a
# The operation takes O(1).
```

```python
# To get Union of two set
a|b
```

```python
# To get Intersection of two set
a&b
```

```python
# To get Difference in two set
a-b
```

```python
# To get symmetic difference between two set
a^b
```

```python
# To check whether a is subset of b
a.issubset(b)
```

```python
# To check whether a is superset of b
a.issuperset(b)
```
