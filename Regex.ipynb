# Special Character
<br>
 A special sequence is a \ followed by one of the characters in the list below, and has a special meaning: .

```python
Image('Special_character.PNG')
```

```python
num_str = "20 managoes costs 20.5 rs "
```

```python
# To extract all the numerical variables
pattern="\d+.\d*"
```

```python
# Find all extract the patterns
re.findall(pattern,num_str)
```

```python
# To find the first match of the pattern
match=re.search(pattern,num_str)
if match:
    print(num_str[match.start():match.end()])
```

```python
# To split based on match
split = re.split(pattern,num_str)
print(split)
```

```python
# To replace the pattern
re.sub(pattern,"<num>",num_str)
```
